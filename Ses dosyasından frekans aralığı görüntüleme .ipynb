```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

# ses doyasını okutma
input_data = read("aaa.wav")
audio = input_data[1]
# 1024'e göre aralığa tanımla ekran boyutu 
plt.plot(audio[0:1024])
# x,y eksenleri adlandırma
plt.ylabel("Frekans genişlik aralığı")
plt.xlabel("Zaman")
# ana başlık adlandırma 
plt.title("Örnek ses dosyası")
# grafiği elde etme
plt.show()
```
