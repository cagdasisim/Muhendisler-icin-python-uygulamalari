```python
import numpy as np
import sympy as sp
import pandas as pd

def solve_beam(l1, l2, q1, q2):
    l=l1+l2 #toplam uzunluk
    Mx=sp.symbols('Mx') #Mx sembolü oluşturulması
    
    #Mx hesabı
    Mx=sp.solveset(Mx*l1/3+q1*l1**3/24+Mx*l2/3+q2*l2**3/24,Mx).args[0]

    #denge denklemleri
    Va, Vb1, Vb2, Vc=sp.symbols('Va Vb1 Vb2 Vc')
    Va, Vb1=sp.linsolve([Va+Vb1-q1*l1, Vb1*l1+Mx-(q1*l1**2)/2], (Va, Vb1)).args[0]
    Vc, Vb2=sp.linsolve([Vb2+Vc-q2*l2, Vb2*l2+Mx-(q2*l2**2)/2], (Vc, Vb2)).args[0]
    Vb=Vb1+Vb2

    x1=np.arange(0,l1+0.1,0.1) #x1 ekseninin oluşturulması
    x2=np.arange(0, l2+0.1, 0.1) #x2 ekseninin oluşturulması

    beam1=pd.DataFrame({"x":x1}) #ilk aralık için bir veri çerçevesi oluşturun
    beam2=pd.DataFrame({"x":x2}) #ikinci aralık için bir veri çerçevesi oluşturun

    beam1["M"]=Va*beam1.x-(q1*beam1.x**2)/2 # M'yi hesapla ve sakla
    beam2["M"]=Mx-(q2*beam2.x**2)/2+Vb2*beam2.x # M'yi hesapla ve sakla

    beam1["V"]=Va-q1*beam1.x # V'yi hesapla ve sakla
    beam2["V"]=Vb2-q2*beam2.x # V'yi hesapla ve sakla

    beam2.x=beam2.x+l1 # ikinci aralık için x'i yeniden atayın

    beam=pd.concat([beam1, beam2]) # iki veri çerçevesini birleştir
    
    return(beam) # sonucu oluştur
```

```python
header=pd.MultiIndex.from_tuples([("combo 1", "M"), ("combo 1", "V"),
                                 ("combo 2", "M"), ("combo 2", "V")])
combos=pd.DataFrame(columns=header)
combos["x"]=solve_beam(4, 5, 3.2, 4.5)["x"]

combos["combo 1"]=solve_beam(4, 5, 3.2, 4.5)
combos["combo 2"]=solve_beam(4, 5, 4.5, 3.2)
combos=combos.set_index("x")

combos=combos.astype("float")
combos.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,5))
ax = plt.subplot(211)
ax.invert_yaxis()

combos.loc[:,pd.IndexSlice[:,"M"]].plot(ax=ax)

ax = plt.subplot(212)
ax.invert_yaxis()
combos.loc[:,pd.IndexSlice[:,"V"]].plot(ax=ax)
fig.savefig("plot.pdf", bbox_inches='tight')
```
