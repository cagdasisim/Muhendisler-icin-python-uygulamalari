```python
from scipy . integrate import odeint
from numpy import *
```

```python
L = 0.46
m = 0.042
x = 0.18
g = 9.81
I0 = m*(L **2/12) + m*x**2
```

```python
t = arange (0 ,5 ,0.05)
def derivadas (state ,t):

  psi,theta = state
  psi_dot = -m*g*x*sin( theta )/I0

  theta_dot = psi
  return [ psi_dot , theta_dot ]

soluciones = odeint ( derivadas , [0, pi /4] ,t)
psi= soluciones [: ,0];theta = soluciones [: ,1]
```

```python
import matplotlib . pyplot as plt

plt.close ('all')
fig = plt. figure (1)
plt.plot (t,theta ,t,psi)
plt.legend (( 'Theta','psi'))
plt.title ('Complex Pandelum')
plt.xlabel ('time,s')
plt.ylabel ('psi,theta,rad')
```
